# **Practical 1 â€” Array Operations**

**(a): CREATE AN ARRAY OF N ELEMENTS**

```python
# Ask user for number of elements
n = int(input("Enter number of elements (N): "))

# Create an empty list (array)
arr = []

# Take input of N integers
print(f"Enter {n} integer elements:")
for i in range(n):
    element = int(input(f"Element {i+1}: "))
    arr.append(element)  # Append each element into array

# Display the array
print("\n Array created successfully!")
print("The elements of the array are:", arr)
```

**(b): INSERT ELEMENT AT A GIVEN VALID POSITION**

```python
# Create initial array
n = int(input("Enter number of elements (N): "))
arr = []

print(f"Enter {n} integer elements:")
for i in range(n):
    element = int(input(f"Element {i+1}: "))
    arr.append(element)

print("\n Initial Array:", arr)

# Ask user for position and new element
pos = int(input("\nEnter position to insert at (0-based index): "))
val = int(input("Enter value to insert: "))

# Validate position
if pos < 0 or pos > len(arr):
    print("Invalid position! Please enter a valid index.")
else:
    # Insert the value
    arr.insert(pos, val)
    print(f"\n Element {val} inserted at position {pos}.")
    print("Updated Array:", arr)
```

**(c): DELETE ELEMENT FROM A GIVEN VALID POSITION**

```python
# Create initial array
n = int(input("Enter number of elements (N): "))
arr = []

print(f"Enter {n} integer elements:")
for i in range(n):
    element = int(input(f"Element {i+1}: "))
    arr.append(element)

print("\n Initial Array:", arr)

# Ask user for position to delete
pos = int(input("\nEnter position to delete (0-based index): "))

# Validate position
if pos < 0 or pos >= len(arr):
    print("Invalid position! Please enter a valid index.")
else:
    # Delete element using pop()
    removed = arr.pop(pos)
    print(f"\n Element {removed} deleted from position {pos}.")
    print("Updated Array:", arr)
```

**Combine all three into one menu-driven program (a+b+c+d)**

```python
# Create Array
def create_array():
    """
    This function creates an array (Python list) of N integer elements.
    """
    n = int(input("Enter number of elements (N): "))  # Take number of elements
    arr = []  # Initialize empty array

    # Take N integer inputs from the user
    for i in range(n):
        element = int(input(f"Enter element {i+1}: "))
        arr.append(element)  # Append element to array

    print("Array created successfully!")
    print("Current Array:", arr)
    return arr
```

```python
# (b) Insert Element
def insert_element(arr):
    """
    This function inserts an element at a given valid position.
    """
    print("\nCurrent Array:", arr)

    # Take position and value from user
    pos = int(input("Enter position to insert at (0-based index): "))
    val = int(input("Enter the value to insert: "))

    # Check if position is valid
    if pos < 0 or pos > len(arr):
        print("Invalid position! Please enter a valid index.")
    else:
        arr.insert(pos, val)  # Insert value at given index
        print(f"Element {val} inserted at position {pos}.")
        print("Updated Array:", arr)

    return arr
```

```python
# (c) Delete Element
def delete_element(arr):
    """
    This function deletes an element from a given valid position.
    """
    print("\nCurrent Array:", arr)

    # Take position to delete from user
    pos = int(input("Enter position to delete (0-based index): "))

    # Check if position is valid
    if pos < 0 or pos >= len(arr):
        print("Invalid position! Please enter a valid index.")
    else:
        removed = arr.pop(pos)  # Delete element using pop()
        print(f"Element {removed} deleted from position {pos}.")
        print("Updated Array:", arr)

    return arr
```

```python
# (d) Exit / Menu System
def main():
    """
    Menu-driven program for array operations.
    """
    arr = []  # Start with an empty array

    while True:
        # Display the menu
        print("\n========= ARRAY OPERATIONS MENU =========")
        print("1. Create Array (a)")
        print("2. Insert Element (b)")
        print("3. Delete Element (c)")
        print("4. Exit (d)")
        print("======================")

        # Take user choice
        choice = int(input("Enter your choice (1-4): "))

        # Perform operations based on choice
        if choice == 1:
            arr = create_array()  # (a)

        elif choice == 2:
            if len(arr) == 0:
                print("Array is empty. Please create an array first.")
            else:
                arr = insert_element(arr)  # (b)

        elif choice == 3:
            if len(arr) == 0:
                print("Array is empty. Nothing to delete.")
            else:
                arr = delete_element(arr)  # (c)

        elif choice == 4:
            print("ðŸ‘‹ Exiting program. Thank you!")  # (d)
            break

        else:
            print("Invalid choice! Please enter a number between 1-4.")


# Run the main function
main()
```
