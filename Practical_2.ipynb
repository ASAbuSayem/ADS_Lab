Part (b1) → Brute Force Pattern Matching

Logic: আমরা মেইন স্ট্রিং-এর প্রতিটি possible অবস্থান থেকে pattern টা চেক করব manually, কোনো built-in find() ব্যবহার না করে।

```python
# (b1): Brute-Force Pattern Matching
def brute_force_search(STR, PAT):
    positions = []
    for i in range(len(STR) - len(PAT) + 1):
        match = True
        for j in range(len(PAT)):
            if STR[i + j] != PAT[j]:
                match = False
                break
        if match:
            positions.append(i)
    return positions

# Testing
positions = brute_force_search(STR, PAT)
if positions:
    print(f"\nPattern found at positions (Brute Force): {positions}")
else:
    print("\nPattern not found (Brute Force)")
```

Part (b2) → Boyer–Moore Algorithm (Simplified Version)

Boyer–Moore অনেক efficient কারণ এটা কিছু character skip করে যায় যখন mismatch হয়।

```python
# (b2): Boyer-Moore (Bad Character Heuristic only)
def boyer_moore_search(STR, PAT):
    m = len(PAT)
    n = len(STR)
    if m == 0:
        return []

    # Create Bad Character Table
    bad_char = [-1] * 256
    for i in range(m):
        bad_char[ord(PAT[i])] = i

    positions = []
    s = 0  # shift of the pattern
    while s <= n - m:
        j = m - 1
        while j >= 0 and PAT[j] == STR[s + j]:
            j -= 1
        if j < 0:
            positions.append(s)
            s += (m - bad_char[ord(STR[s + m])] if s + m < n else 1)
        else:
            s += max(1, j - bad_char[ord(STR[s + j])])
    return positions

# Testing
positions_bm = boyer_moore_search(STR, PAT)
if positions_bm:
    print(f"Pattern found at positions (Boyer-Moore): {positions_bm}")
else:
    print("Pattern not found (Boyer-Moore)")
```

Part (b3) → Knuth-Morris-Pratt (KMP) Algorithm

এটা pattern এর নিজস্ব structure ব্যবহার করে unnecessary comparisons এড়ায়।

```python
# (b3): KMP Algorithm
def compute_lps(PAT):
    lps = [0] * len(PAT)
    length = 0
    i = 1
    while i < len(PAT):
        if PAT[i] == PAT[length]:
            length += 1
            lps[i] = length
            i += 1
        else:
            if length != 0:
                length = lps[length - 1]
            else:
                lps[i] = 0
                i += 1
    return lps

def kmp_search(STR, PAT):
    lps = compute_lps(PAT)
    i = j = 0
    positions = []
    while i < len(STR):
        if PAT[j] == STR[i]:
            i += 1
            j += 1
        if j == len(PAT):
            positions.append(i - j)
            j = lps[j - 1]
        elif i < len(STR) and PAT[j] != STR[i]:
            if j != 0:
                j = lps[j - 1]
            else:
                i += 1
    return positions

# Testing
positions_kmp = kmp_search(STR, PAT)
if positions_kmp:
    print(f"Pattern found at positions (KMP): {positions_kmp}")
else:
    print("Pattern not found (KMP)")
```
