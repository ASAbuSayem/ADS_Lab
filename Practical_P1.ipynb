**Practical 1: Design, Develop and Implement a menu driven program in python for the following ARRAY operations**

```python
# a: Create an Array of N Integer Elements
def create_array():
    n = int(input("Enter number of elements (N): "))
    arr = []
    print(f"Enter {n} integer elements:")
    for i in range(n):
        element = int(input(f"Element {i+1}: "))
        arr.append(element)
    print("\nArray created successfully!")
    print("Current Array:", arr)
    return arr
```

```python
# b: Insert an Element at a Given Valid Position
def insert_element(arr):
    if len(arr) == 0:
        print("Array is empty! Please create an array first.")
        return arr

    print("\nCurrent Array:", arr)
    pos = int(input("Enter position to insert at (0-based index): "))
    val = int(input("Enter value to insert: "))

    if pos < 0 or pos > len(arr):
        print("Invalid position! Please enter a valid index.")
    else:
        arr.insert(pos, val)
        print(f"\nElement {val} inserted at position {pos}.")
        print("Updated Array:", arr)
    return arr
```

```python
# Delete an Element from a Given Valid Position
def delete_element(arr):
    if len(arr) == 0:
        print("Array is empty! Nothing to delete.")
        return arr

    print("\nCurrent Array:", arr)
    pos = int(input("Enter position to delete (0-based index): "))

    if pos < 0 or pos >= len(arr):
        print("Invalid position! Please enter a valid index.")
    else:
        removed = arr.pop(pos)
        print(f"\nElement {removed} deleted from position {pos}.")
        print("Updated Array:", arr)
    return arr
```

```python
# d: Main Menu/Exit (Menu-driven program for array operations)
def main():
    arr = []

    while True:
        print("\n===== ARRAY OPERATIONS MENU =====")
        print("1. Create Array (a)")
        print("2. Insert Element (b)")
        print("3. Delete Element (c)")
        print("4. Exit (d)")
        print("======================")

        choice = int(input("Enter your choice (1-4): "))

        if choice == 1:
            arr = create_array()
        elif choice == 2:
            arr = insert_element(arr)
        elif choice == 3:
            arr = delete_element(arr)
        elif choice == 4:
            print("\nðŸ‘‹ Exiting Program. Thank You!")
            break
        else:
            print("Invalid choice! Please enter a number between 1â€“4.")

main()
```
