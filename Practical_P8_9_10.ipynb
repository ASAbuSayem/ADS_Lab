**8. WAP on Linear Queue – Insertion , Deletion , Display**

```python
class LinearQueue:
    def __init__(self, size):
        self.size = size
        self.queue = [None] * size
        self.front = -1
        self.rear = -1

    # Insertion (Enqueue)
    def enqueue(self, elements):
        for elem in elements:
            if self.rear == self.size - 1:
                print("Queue Overflow! Cannot insert more elements.")
                return
            if self.front == -1:
                self.front = 0
            self.rear += 1
            self.queue[self.rear] = elem
            print(f"Inserted: {elem}")

    # Deletion (Dequeue)
    def dequeue(self):
        if self.front == -1 or self.front > self.rear:
            print("Queue Underflow! No elements to delete.")
            return
        print(f"Deleted: {self.queue[self.front]}")
        self.front += 1
        if self.front > self.rear:
            self.front = self.rear = -1

    # Display Elements
    def display(self):
        if self.front == -1:
            print("Queue is Empty!")
        else:
            print("\nCurrent Queue Elements:")
            for i in range(self.front, self.rear + 1):
                print(self.queue[i], end="  ")
            print("\nFront =", self.front, " Rear =", self.rear)
```

```python
# Main Program
if __name__ == "__main__":
    size = int(input("Enter the size of the Queue: "))
    q = LinearQueue(size)

    while True:
        print("\n===== MENU =====")
        print("1. Insert (Enqueue)")
        print("2. Delete (Dequeue)")
        print("3. Display Queue")
        print("4. Exit")
        print("===================")

        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            elements = input("Enter elements to insert (separated by space): ").split()
            q.enqueue(elements)

        elif choice == '2':
            q.dequeue()

        elif choice == '3':
            q.display()

        elif choice == '4':
            print("Exiting Program... Goodbye!")
            break

        else:
            print("Invalid choice! Please try again.")
```

**9. WAP on Double Queue – Insertion , Deletion , Display**

```python
class Deque:
    def __init__(self, size):
        self.size = size
        self.queue = []

    def insert_front(self, item):
        if len(self.queue) == self.size:
            print("Queue is full! Cannot insert at front.")
        else:
            self.queue.insert(0, item)
            print(f"{item} inserted at FRONT.")

    def insert_rear(self, item):
        if len(self.queue) == self.size:
            print("Queue is full! Cannot insert at rear.")
        else:
            self.queue.append(item)
            print(f"{item} inserted at REAR.")

    def delete_front(self):
        if not self.queue:
            print("Queue is empty! Nothing to delete from front.")
        else:
            item = self.queue.pop(0)
            print(f"{item} deleted from FRONT.")

    def delete_rear(self):
        if not self.queue:
            print("Queue is empty! Nothing to delete from rear.")
        else:
            item = self.queue.pop()
            print(f"{item} deleted from REAR.")

    def display(self):
        if not self.queue:
            print("Queue is empty!")
        else:
            print("\n Current Queue Elements:")
            print(" <--> ".join(map(str, self.queue)))
            print(f"Front = {self.queue[0]}, Rear = {self.queue[-1]}\n")
```

```python
print("===== DOUBLE ENDED QUEUE (DEQUE) =====")
size = int(input("Enter the maximum size of queue: "))
dq = Deque(size)

while True:
    print("\n===== MENU =====")
    print("1. Insert at Front")
    print("2. Insert at Rear")
    print("3. Delete from Front")
    print("4. Delete from Rear")
    print("5. Display Queue")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")

    if choice == '1':
        val = input("Enter value to insert at FRONT: ")
        dq.insert_front(val)
    elif choice == '2':
        val = input("Enter value to insert at REAR: ")
        dq.insert_rear(val)
    elif choice == '3':
        dq.delete_front()
    elif choice == '4':
        dq.delete_rear()
    elif choice == '5':
        dq.display()
    elif choice == '6':
        print("Exiting Program. Bye Bye")
        break
    else:
        print("Invalid choice! Please enter 1–6.")
```

**10. WAP on Circular Queue – Insertion , Deletion , Display**

```python
class CircularQueue:
    def __init__(self, size):
        self.size = size
        self.queue = [None] * size
        self.front = -1
        self.rear = -1

    # INSERTION
    def enqueue(self, item):
        if (self.rear + 1) % self.size == self.front:
            print("Queue is FULL! Cannot insert.")
        elif self.front == -1:  # Empty queue
            self.front = 0
            self.rear = 0
            self.queue[self.rear] = item
            print(f"Inserted {item} at position {self.rear}.")
        else:
            self.rear = (self.rear + 1) % self.size
            self.queue[self.rear] = item
            print(f"Inserted {item} at position {self.rear}.")

    # DELETION
    def dequeue(self):
        if self.front == -1:
            print("Queue is EMPTY! Nothing to delete.")
        elif self.front == self.rear:
            print(f"Deleted {self.queue[self.front]} from position {self.front}.")
            self.queue[self.front] = None
            self.front = -1
            self.rear = -1
        else:
            print(f"Deleted {self.queue[self.front]} from position {self.front}.")
            self.queue[self.front] = None
            self.front = (self.front + 1) % self.size

    # DISPLAY
    def display(self):
        if self.front == -1:
            print("Queue is EMPTY!")
        else:
            print("\n Current Circular Queue Elements:")
            if self.rear >= self.front:
                for i in range(self.front, self.rear + 1):
                    print(self.queue[i], end="  ")
            else:
                for i in range(self.front, self.size):
                    print(self.queue[i], end="  ")
                for i in range(0, self.rear + 1):
                    print(self.queue[i], end="  ")
            print(f"\nFront = {self.front}, Rear = {self.rear}\n")
```

```python
print("===== CIRCULAR QUEUE IMPLEMENTATION =====")
size = int(input("Enter the maximum size of queue: "))
cq = CircularQueue(size)

while True:
    print("\n===== MENU =====")
    print("1. Insert (Enqueue)")
    print("2. Delete (Dequeue)")
    print("3. Display Queue")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        val = input("Enter value to insert: ")
        cq.enqueue(val)
    elif choice == '2':
        cq.dequeue()
    elif choice == '3':
        cq.display()
    elif choice == '4':
        print("Exiting Program...")
        break
    else:
        print("Invalid choice! Please enter 1–4.")
```
