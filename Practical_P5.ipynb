```python
# PRACTICAL 5: HASHING (LINEAR PROBING)
# Function to insert a key using linear probing
def insert(hash_table, key, m):
    # Calculate initial hash address using remainder method
    hash_address = key % m
    print(f"\nKey {key} hashed to index {hash_address}")

    # If slot is empty, insert directly
    if hash_table[hash_address] == -1:
        hash_table[hash_address] = key
        print(f"Inserted {key} at index {hash_address}")
    else:
        print(f"Collision detected at index {hash_address} for key {key}")
        # Linear probing: find next empty slot
        i = (hash_address + 1) % m
        while hash_table[i] != -1 and i != hash_address:
            print(f"Slot {i} already occupied by {hash_table[i]}, checking next...")
            i = (i + 1) % m
        # Insert if an empty slot is found
        if hash_table[i] == -1:
            hash_table[i] = key
            print(f"Inserted {key} at index {i}")
        else:
            print(f"Hash Table is full! Could not insert key {key}.")
```

```python
# Function to display hash table
def display_table(hash_table):
    print("\nFinal Hash Table:")
    print("-----------------")
    for i, key in enumerate(hash_table):
        if key == -1:
            print(f"Slot {i}: Empty")
        else:
            print(f"Slot {i}: {key}")
```

```python
# MAIN PROGRAM
# Take user inputs
m = int(input("Enter the size of hash table (m): "))
n = int(input("Enter number of employee records (n): "))

# Initialize hash table with -1 (empty slots)
hash_table = [-1] * m

# Insert N keys into the hash table
print("\nEnter employee keys (4-digit integers):")
for _ in range(n):
    key = int(input("Enter key: "))
    insert(hash_table, key, m)

# Display final hash table
display_table(hash_table)
```
