**Practical 12: WAP to implement insertion , deletion on stack.**

```python
stack = []
MAX_SIZE = 5
```

```python
# Function to insert (PUSH)
def push():
    if len(stack) == MAX_SIZE:
        print("\n Stack Overflow! No more elements can be inserted.")
    else:
        element = input("Enter element to PUSH: ")
        stack.append(element)
        print(f" {element} inserted (PUSHED) into stack.")
```

```python
# Function to delete (POP)
def pop_element():
    if len(stack) == 0:
        print("\n Stack Underflow! Stack is empty.")
    else:
        removed = stack.pop()
        print(f" {removed} deleted (POPPED) from stack.")
```

```python
# Function to display stack
def display():
    if len(stack) == 0:
        print("\n Stack is empty.")
    else:
        print("\n Current Stack (TOP → BOTTOM):")
        for item in reversed(stack):
            print(item)
```

```python
# MENU
while True:
    print("\n===== STACK MENU =====")
    print("1. PUSH (Insert)")
    print("2. POP (Delete)")
    print("3. DISPLAY")
    print("4. EXIT")

    choice = input("Enter your choice: ")

    if choice == '1':
        push()
    elif choice == '2':
        pop_element()
    elif choice == '3':
        display()
    elif choice == '4':
        print("\nExiting program...")
        break
    else:
        print("\n Invalid Choice! Please select 1–4.")
```
